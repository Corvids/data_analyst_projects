## Age, Class, and Sex

Intuitively, I would assume that survival rate would vary by passenger class (__pclass__), age (__Age__), and gender (__sex__).

Let's first look at at the distribution of survivors by __Age__:

```python
sns.stripplot(x="Survived", y="Age", data=df, jitter=True);
```

```python
sns.boxplot(x="Survived", y="Age", hue="Sex", palette="pastel", data=df);
```

Let's look at the same graph but with __Fare__ replacing __Age__:

```python
sns.stripplot(x="Survived", y="Fare", data=df, jitter=True);
```

```python
sns.boxplot(x="Survived", y="Fare", hue="Sex", palette="pastel", data=df);
```

Looking at a combination of __Age__ and __Fare__, it looks like younger passengers who paid a higher fare were somewhat more likely to survive:

```python
sns.lmplot(x="Age", y="Fare", hue="Survived", data=df,
           markers=["o", "x"], palette="Set1");
```
